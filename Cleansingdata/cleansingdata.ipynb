```python
import numpy as np
import pandas as pd
```

```python
#reading CSV _1 
suicide_gender_ratios = pd.read_csv("Male-Female-Ratio-of-Suicide-Rates.csv").sort_values(by='Entity', ascending=True)
```

```python
#reading CSV _2
suicide_all = pd.read_csv("suicide-rates-by-country.csv").sort_values(by='Entity', ascending=True)
```

```python
#merging both CSV files
sui_merge = pd.merge(suicide_gender_ratios, suicide_all, how='inner').sort_values(by='Entity', ascending=True)
sui_merge.columns = ['Country', 'Code', 'Year', 'Male-Female ratio', 'Total suicide rate']
```

```python
sui_merge.head()
```

```python
#counting merged files
sui_merge.count()
```

```python
#drop NA
sui_merge= sui_merge.dropna(how='any')
sui_merge.count()
```

```python
#rename columns
cleaning_data=sui_merge.rename(columns={'Code':'Country Code'})
cleaning_data.head(50)
```

```python
cleaning_data.to_csv('cleanDATA.csv',index= False)
```

```python
#cleaning the data- and finding mean for every country by year

groupby_1 = cleaning_data.groupby(['Year', 'Country']).agg({'Male-Female ratio': 'mean', 'Total suicide rate': 'mean'})
groupby_1.head()
```

```python
#adding third CSV file to see if there are any correlations between suicide and poverty levels
poverty_data = pd.read_csv("subnational_poverty_country.csv").sort_values(by='country_name', ascending=True)
poverty_data.head()
```

```python
#removing columns
poverty_cleaned = poverty_data.drop(columns=['indicator_name','Unnamed: 2','Unnamed: 3','Unnamed: 4','country_code','year'])
poverty_cleaned.head()
```

```python
#renaming columns
poverty_cleaned_new=poverty_cleaned.rename(columns={'country_name':'Country','value':'Poverty Value'})

poverty_cleaned_new.head()
```

```python
#drop NA
poverty_new= poverty_cleaned_new.dropna(how='any')
poverty_new.count()
```

```python
#merging third CSV

merge_all = pd.merge(cleaning_data, poverty_new, on="Country", how="left")
merge_all.head()
```

```python
#cleaning the data- and finding mean for every country by year AND including Poverty Values

groupby_2 = merge_all.groupby(['Year', 'Country']).agg({'Male-Female ratio': 'mean', 'Total suicide rate': 'mean','Poverty Value':'mean'})
groupby_2.head(200)
```

```python
groupby_2.to_csv('ETLproject.csv',index= False)
```
